```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('/content/Real_Estate_Sales_2001-2022_GL-Short.csv', index_col="Serial Number") # Added the filepath to the read_csv function
print(df)
```

```python
print(df.info())
print(df.dtypes)
print(df.describe())
print(df.shape)
```

# **Plots**

3) Line plot


```python
sns.lineplot(data=df, x="Town", y="Assessed Value")
plt.xticks(rotation=90)
plt.title("Line – Town vs Assessed Value")
plt.show()
```

4) Categorical plot (defaults to bar)

```python
sns.catplot(data=df, x="Town", y="Assessed Value", kind="bar", height=4, aspect=2)
plt.xticks(rotation=90)
plt.title("CatPlot – Town vs Assessed Value")
plt.show()
```

```python
for categorical vs continuous data
sns.violinplot(data=df, x="Town", y="Assessed Value")
plt.xticks(rotation=90)
plt.title("Violin Plot – Town vs Assessed Value")
plt.show()
```

6) Scatter plot

```python
sns.scatterplot(data=df, x="Town", y="Assessed Value")
plt.xticks(rotation=90)
plt.title("Scatter – Town vs Assessed Value")
plt.show()
```

 7) Bar plot

```python
sns.barplot(data=df, x="Town", y="Assessed Value", estimator="mean")
plt.xticks(rotation=90)
plt.title("Bar – Town vs Assessed Value (mean)")
plt.show()
```

8) Heat-map (pivot first: Town rows, 1 column of mean values)


```python
town_av = df.pivot_table(values="Assessed Value", index="Town", aggfunc="mean")
sns.heatmap(town_av, cmap="Blues")
plt.title("Heatmap – Town vs Assessed Value (mean)")
plt.show()
```

27 & 28 – Trying Seaborn themes

```python
themes = ["darkgrid", "whitegrid", "dark", "white", "ticks"]

# 27) Five line plots with different themes
for style in themes:
    sns.set_theme(style=style)
    sns.lineplot(data=df, x="Town", y="Assessed Value")
    plt.xticks([], [])         # hide crowded x-labels for cleaner view
    plt.title(f"Theme: {style} (line plot)")
    plt.show()

# 28) Five bar plots with the same 5 styles
for style in themes:
    sns.set_style(style)
    sns.barplot(data=df, x="Property Type", y="Sale Amount", estimator="mean")
    plt.xticks([], [])
    plt.title(f"Theme: {style} (bar plot)")
    plt.show()
```

29 – Your own custom theme

```python
custom_rc = {
    "axes.facecolor": "#EAEAF2",
    "axes.edgecolor": "red",
    "axes.grid": True,
    "grid.color": "green",
    "grid.linewidth": 0.7,
    "axes.titlesize": "large",
    "axes.labelsize": "medium",
    "font.size": 10,
    "figure.figsize": (8, 4)
}

sns.set_theme(style="whitegrid", rc=custom_rc)

# test the custom theme on any 5 plots you like:
for i in range(5):
    sns.lineplot(data=df, x="Year", y="Assessed Value")
    plt.title(f"Custom theme example #{i+1}")
    plt.show()
```
